```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
df = pd.read_csv('F:/Git/deep_learning/data/placement.csv')
```

```python
df.head()
```

```python
df.drop('si.no', axis=1, inplace=True)
```

```python
df.head()
```

```python
sns.scatterplot(x='cgpa', y='iq', hue='placement', data=df)
plt.show()
```

```python
x= df.iloc[:, :-1]
y=df["placement"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=20)
```

```python
from sklearn.linear_model import Perceptron
```

```python
pr = Perceptron(alpha=0.01)
pr.fit(x_train, y_train)
```

```python
pr.score(x_test, y_test)*100, pr.score(x_train, y_train)*100
```

```python
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=pr)
plt.show()
```
